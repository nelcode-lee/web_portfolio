## Standardise Column values 

#### Hits Column - fix string and empty cells

```python
def clean_hits (x):
    if pd.isna(x):
        return 0
    elif 'K' in str(x):
        return float(x[:-1])*1000
    else:
        return float(x) 

df['Hits_K'] = df['Hits'].apply(clean_hits)

df.drop(columns=['Hits'], inplace=True)

print(df['Hits_K'].head(10))
```

```python
df['Hits_K'] = df['Hits_K'].astype(int)
print(df['Hits_K'].head(10))
```

#### Height standardisation in cm

```python
# clean height column, remove cm
df['Height'].head()
```

```python
df['Height'].unique()
```

```python
df['Height'] = df['Height'].str.strip('cm')
df['Height'].unique()
```

```python
# standardise height measurement to cm (conversion = d(cm) = 2ft × 30.48 + 3in × 2.54 = 68.58cm)

def ft_to_cm(x):
    if "'" in x:
        ft=x.replace("'",".").replace('"',"")
        cm= round(float(ft)*30.48,0)
        return int(cm)
    else:
        return int(x)

df["Height"]=df["Height"].apply(ft_to_cm)
```

```python
df = df.rename(columns={'Height': 'Height_cm'})
```

```python
df['Height_cm'].unique()
```

#### Weight standardisation in kg

```python
df['Weight'] = df['Weight'].str.strip('kg')
df = df.rename(columns={'Weight': 'Weight_kg'})
df.head()
```

```python
df['Weight_kg'].unique()
```

```python
# standardise value to kgs - ( 1 kg = 2.205 lb)

def lbs_to_kg(x):
    if 'lbs' in x:
        lbs=x.replace('lbs','')
        kgs= round(int(lbs)/2.2,0)
        return int(kgs)
    else:
        return int(x)

df['Weight_kg']=df['Weight_kg'].apply(lbs_to_kg)
```

#### Standardise financial values 

```python
# Value, Wage & Release Clause - standardise 
df['Value'].unique()
```

```python
df[['Value', 'Wage', 'Release Clause']].head()
```

```python
def cash(x):
    if "€" in x:
        x = x.replace('€', '')

    if 'M' in x:
        x= x.replace('M', '')
        return int(float(x)*1000000)

    elif 'K' in x:
        x= x.replace('K', '')
        return int(float(x)*1000)

df['Value'] = df['Value'].apply(cash) / 1_000_000
df['Wage'] = df['Wage'].apply(cash)
df['Release Clause'] = df['Release Clause'].apply(cash) / 1_000_000
        
df.rename(columns={
    "Value": 'Values (M)',
    "Wage": 'Wage in € ',
    "Release Clause": "Release Clause in M€"
}, inplace=True)  
```

#### Standardise Contract dates 

```python
df ['Contract'].unique()
```

```python
from datetime import datetime 
```

```python
def type(x):
    if "Free" in x:
        return "Free"
    if "Loan" in x:
        return "Loan"
    if "~" in x:
        return "Contract"
    else:
        return pd.na


def start_time_contract(x):
    if "~" in x:
        return  int(x[:4])
    if "Loan" in x:
        x=x.strip(" On Loan")
        x= datetime.strptime(x, "%b %d, %Y")
        return x.date()
    else:
        return pd.NA

def end_time_contract(type,contract,loan):
    if type == "Contract":
        return int(contract[-4:])
    if type == "Loan":
        loan = datetime.strptime(loan, "%b %d, %Y")
        return loan.date()

    else:
        return pd.NA

df["Type of contract"] = df["Contract"].apply(type)
df["Start year"] = df["Contract"].apply(start_time_contract)
df["End year"] = df.apply(lambda row: end_time_contract(row["Type of contract"],row["Contract"],row["Loan Date End"] ), axis=1)
```
